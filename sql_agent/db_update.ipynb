### 업데이트


```python
import duckdb

con = duckdb.connect("db/manufacturing.db")

con.execute("ALTER TABLE fact_manufacturing ADD COLUMN IF NOT EXISTS ingest_order BIGINT;")
con.execute("UPDATE fact_manufacturing SET ingest_order = rowid WHERE ingest_order IS NULL;")
con.execute("CREATE INDEX IF NOT EXISTS idx_fact_ingest ON fact_manufacturing(ingest_order);")

print("✅ ingest_order 컬럼이 추가 및 업데이트 완료되었습니다.")
```
